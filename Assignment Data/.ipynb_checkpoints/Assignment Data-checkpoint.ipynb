```python
rage_crime_plot = sum_of_offences['Harassment, threatening behaviour and private nuisance'].plot()
```

Nice. Now we overlay the release dates of the selected video games. Here are our release dates:

```python
for i in range(len(game_data)) : 
  print(game_data.loc[i, "Video Game"], game_data.loc[i, "Release Date"]) 

game_data.loc[i, "Video Game"]
list_of_release_dates = game_data['Release Date'].values
list_of_release_dates
```

To plot our release dates as vertical lines we'll need the entries as indices in sum_of_offences. First lets select

```python
rage_crime_df = sum_of_offences['Harassment, threatening behaviour and private nuisance']
```

```python
list_of_indices = []
for rdate in list_of_release_dates:
    list_of_indices.append(    rage_crime_df.index.get_loc(rdate))
list_of_indices
```

```python
axes = sum_of_offences['Harassment, threatening behaviour and private nuisance'].plot()
ymin, ymax, = axes.get_ylim()
axes.vlines(x=list_of_indices, ymin=ymin, ymax=ymax-1, color='b')
game_title_annotation = "annotate"

for i in range(0,len(list_of_indices)):
    axes.annotate(game_data["Video Game"].values[i],
              xy=(list_of_indices[i], ymax/3),
                #xy=(list_of_indices[i]-((i%2)*2.5), 100+((i%2)*150)),
              xycoords='data',
              xytext=(-75,-100+((i%2)*30)),
              textcoords='offset points',
              arrowprops=dict(headwidth=10, width=4, color='#363d46', connectionstyle="angle3,angleA=0,angleB=-90"),
              fontsize=12)
```

As you can see, the general level of rage in society has been steadility increasing since 1995. This growth however oscillates more widely when we start to look at the game releases in our data range.

Looking carefully you can see a consistent trend in data spikes that follow the initial release dates.
